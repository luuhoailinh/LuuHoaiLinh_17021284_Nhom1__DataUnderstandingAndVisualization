## dataset: all_data.csv
    
Description: Life expectancy and gdp of 6 countries

```python
from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
#load data

df = pd.read_csv("all_data.csv")
```

```python
#some info about data

df.info()
df.head()
```

```python
#name of countries

unique_countries = list(df['Country'].unique())
unique_countries
```

```python
#represented years

unique_years = df['Year'].unique()
unique_years
```

```python
#rename col

df = df.rename(columns = {'Life expectancy at birth (years)':'LEABY'})

#view
df.head()
```

## bar chart: GDP of countries

```python
fig, ax = plt.subplots(figsize = (12, 8))

sns.barplot(data = df, x = 'Country', y = 'GDP')

plt.title('GDP by Country (2000-2015)')

plt.xlabel('Countries')
plt.ylabel('GDP in Trillions of US Dollars')

#thay đổi khoảng giá trị
# ax.set_xticks(range(6))

#decor for x-axe
# ax.set_xticklabels(unique_countries, rotation = 45)

plt.show()
```

## bar chart: Life expectancy at birth of countries

```python
fig, ax = plt.subplots(figsize = (12, 8))

sns.barplot(data = df, x = 'Country', y = 'LEABY')

plt.title('Life expectancy at birth (2000-2015)')

plt.xlabel('Countries')
plt.ylabel('LEABY')

#thay đổi khoảng giá trị
# ax.set_xticks(range(6))

#decor for x-axe
# ax.set_xticklabels(unique_countries, rotation = 45)

plt.show()
```

## Bar plot over time

```python
sns.set_style('darkgrid')

f, ax = plt.subplots(figsize=(15, 15)) 

#hue for years
ax = sns.barplot(data=df, x='Country', y='GDP', hue='Year')

plt.title('GDP and Life expectancey over time')
ax.set_xticklabels(unique_countries, rotation = 45)
plt.show()
```

```python
sns.set_style('darkgrid')

f, ax = plt.subplots(figsize=(15, 15)) 
ax = sns.barplot(data=df, x='Country', y='LEABY', hue='Year')
plt.title('GDP and Life expectancey over time')
ax.set_xticklabels(unique_countries, rotation = 45)
ax.set(ylabel = 'Life expectancy at birth')
plt.show()
```
